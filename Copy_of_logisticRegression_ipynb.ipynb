```python
import pandas as pa
import numpy as nu
import matplotlib.pyplot as mat
import seaborn as sea
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings('ignore')
from google.colab import drive
drive.mount("/content/drive")
rds=pa.read_csv("/content/winequality-red.csv")
rds.shape
rds.head()
rds.isnull().sum()
rds.describe()
sea.catplot(x='quality',data=rds,kind="count")
a=mat.figure(figsize=(6,6))
mat.title("quality vs fixed acidity",fontsize=30)
sea.barplot(x="quality",y="fixed acidity",data=rds)
sea.barplot(x="quality",y="density",data=rds)
b=mat.figure(figsize=(6,6))
mat.title("quality vs volatile acidity",fontsize=30)
sea.barplot(x="quality",y="volatile acidity",data=rds)
c=mat.figure(figsize=(6,6))
mat.title("quality vs citric acid",fontsize=30)
sea.barplot(x="quality",y="citric acid",data=rds)
d=mat.figure(figsize=(6,6))
mat.title("quality vs residual sugar",fontsize=30)
sea.barplot(x="quality",y="residual sugar",data=rds)
e=mat.figure(figsize=(6,6))
mat.title("quality vs chlorides",fontsize=30)
sea.barplot(x="quality",y="chlorides",data=rds)
f=mat.figure(figsize=(6,6))
mat.title("quality vs free sulfur dioxide",fontsize=30)
sea.barplot(x="quality",y="free sulfur dioxide",data=rds)
g=mat.figure(figsize=(6,6))
mat.title("quality vs total sulfur dioxide",fontsize=30)
sea.barplot(x="quality",y="total sulfur dioxide",data=rds)
h=mat.figure(figsize=(6,6))
mat.title("quality vs density",fontsize=30)
sea.barplot(x="quality",y="density",data=rds)
i=mat.figure(figsize=(6,6))
mat.title("quality vs pH",fontsize=30)
sea.barplot(x="quality",y="pH",data=rds)
j=mat.figure(figsize=(6,6))
mat.title("quality vs sulphates",fontsize=30)
sea.barplot(x="quality",y="sulphates",data=rds)
k=mat.figure(figsize=(6,6))
mat.title("quality vs alcohol",fontsize=30)
sea.barplot(x="quality",y="alcohol",data=rds)
correlation=rds.corr()
mat.figure(figsize=(12,12))
sea.heatmap(correlation , cbar=True , square=True , fmt='0.1f' , annot=True ,annot_kws={'size':8} , cmap='Greens')
x=rds.drop('quality',axis=1)
print(x)
y=rds['quality'].apply(lambda b_value:'high quality' if b_value>=6 else 'low quality')
print(y)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
print(y.shape,y_train.shape,y_test.shape)
model1=LogisticRegression()
model1.fit(x_train,y_train)
x_test_prediction=model1.predict(x_test)
test_accuracy=accuracy_score(x_test_prediction,y_test)
print("accuracy:",test_accuracy*100)
```
